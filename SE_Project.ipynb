```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Data and Drop Columns

```python
nutrient_data = pd.read_excel('/home/anna/Desktop/Stream Ecology/Ecel Sheets/Antarctic_Season1_CNP_Enzymes_Metadata_no_formulas.xlsx')
nutrient_data.columns
# nutrient_data.head()
nutrient_data = nutrient_data.drop(columns=['Wet_dry_conversion', 'Qubit_raw_ng/uL','g_wet_extracted_DNA', 'Qubit_ng_DNA/g_cryo',
       'Chl-a_raw_ug/L', 'g_wet_extracted_chla', 'ug_chl-a/g_cryo',
       'g_weighboat', 'g_wet', 'g_dry_plusboat', 'g_ashed_plusboat', 'Tardigrades_extracted', 'Rotifers_extracted',
       'Ciliates_extracted', 'ml_in_extraction', 'ml_counted', 
      'TC_percent', 'TN_percent', 'BG/(NAG+LAP)', '(NAG+LAP)/Phos',
       'BG/Phos', 'AG', 'BG', 'BXYL', 'CBH', 'LAP', 'NAG', 'PHOS',
       'percent_OM_nonacifidified', 'percent_OM_acid',
       'percent_OM_leached_with_water_instead_of_acid'])
nutrient_data.head()
```

### Canada Glacier Filling in Missing Values using the Mean of Each Column

```python
caa = nutrient_data.iloc[:30]
caa_columns = caa.columns
print(caa_columns[0])


per_org_mean = caa['percent_organic'].mean()
f_mean = caa['Fluorescence'].mean()
microfauna_mean = caa['g_wet_microfauna'].mean()
tard_mean = caa['Tardigrades/g_dry'].mean()
rot_mean = caa['Rotifers/g_dry'].mean()
cil_mean = caa['Ciliates/g_dry'].mean()
ph_mean = caa['pH_lab'].mean()
elec_mean = caa['electrical_conductivity'].mean()

caa['percent_organic'].fillna(per_org_mean, inplace=True)
caa['Fluorescence'].fillna(f_mean, inplace=True)
caa['g_wet_microfauna'].fillna(microfauna_mean, inplace=True)
caa['Tardigrades/g_dry'].fillna(tard_mean, inplace=True)
caa['Rotifers/g_dry'].fillna(rot_mean, inplace=True)
caa['Ciliates/g_dry'].fillna(cil_mean, inplace=True)
caa['pH_lab'].fillna(ph_mean, inplace=True)
caa['electrical_conductivity'].fillna(elec_mean, inplace=True)
```

### Commonwealth Glacier Filling in Missing Values using the Mean of Each Column

```python
coo = nutrient_data.iloc[30:60]
# print(coo['SampleID'])


per_org_mean = coo['percent_organic'].mean()
f_mean = coo['Fluorescence'].mean()
microfauna_mean = coo['g_wet_microfauna'].mean()
tard_mean = coo['Tardigrades/g_dry'].mean()
rot_mean = coo['Rotifers/g_dry'].mean()
cil_mean = coo['Ciliates/g_dry'].mean()
ph_mean = coo['pH_lab'].mean()
elec_mean = coo['electrical_conductivity'].mean()

coo['percent_organic'].fillna(per_org_mean, inplace=True)
coo['Fluorescence'].fillna(f_mean, inplace=True)
coo['g_wet_microfauna'].fillna(microfauna_mean, inplace=True)
coo['Tardigrades/g_dry'].fillna(tard_mean, inplace=True)
coo['Rotifers/g_dry'].fillna(rot_mean, inplace=True)
coo['Ciliates/g_dry'].fillna(cil_mean, inplace=True)
coo['pH_lab'].fillna(ph_mean, inplace=True)
coo['electrical_conductivity'].fillna(elec_mean, inplace=True)
print(coo)
```

### Taylor Glacier Filling in Missing Values using the Mean of Each Column

```python
tay = nutrient_data.iloc[60:90]
# print(tay['SampleID'])


per_org_mean = tay['percent_organic'].mean()
f_mean = tay['Fluorescence'].mean()
microfauna_mean = tay['g_wet_microfauna'].mean()
tard_mean = tay['Tardigrades/g_dry'].mean()
rot_mean = tay['Rotifers/g_dry'].mean()
cil_mean = tay['Ciliates/g_dry'].mean()
ph_mean = tay['pH_lab'].mean()
elec_mean = tay['electrical_conductivity'].mean()

tay['percent_organic'].fillna(per_org_mean, inplace=True)
tay['Fluorescence'].fillna(f_mean, inplace=True)
tay['g_wet_microfauna'].fillna(microfauna_mean, inplace=True)
tay['Tardigrades/g_dry'].fillna(tard_mean, inplace=True)
tay['Rotifers/g_dry'].fillna(rot_mean, inplace=True)
tay['Ciliates/g_dry'].fillna(cil_mean, inplace=True)
tay['pH_lab'].fillna(ph_mean, inplace=True)
tay['electrical_conductivity'].fillna(elec_mean, inplace=True)
print(tay)
```

### Putting the entire dataframe back together

```python
cleaned = pd.concat([caa, coo, tay], ignore_index=True)
print(cleaned.columns)
```
